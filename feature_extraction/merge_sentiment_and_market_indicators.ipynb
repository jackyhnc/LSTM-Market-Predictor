```python
import pandas as pd

df_market_features = pd.read_csv('market_features_no_sentiment.csv')
df_sentiment = pd.read_csv('sentiment.csv').rename(columns={'date': 'Date'})
```

```python
# intersection dates
for date in df_market_features["Date"]:
  if not df_sentiment["Date"].str.contains(date).any():
    df_market_features = df_market_features.drop(df_market_features[df_market_features["Date"] == date].index)

df_market_features = df_market_features.reset_index(drop=True)
```

```python
print(df_sentiment.shape)
print(df_market_features.shape)
```

```python
tickers = set(col.split("_")[0] for col in df_market_features.columns)

for ticker in tickers:
  if ticker not in df_sentiment.columns:
    continue

  df_market_features[f"{ticker}_Sentiment"] = df_sentiment[ticker]
  if df_market_features[f"{ticker}_Sentiment"].isna().any():
    print(ticker)
```

```python
sentiment_cols = [col for col in df_market_features.columns if col.endswith('_Sentiment')]
df_market_features[sentiment_cols] = df_market_features[sentiment_cols].replace({
    'neutral': 0,
    'positive': 1,
    'negative': -1
})
df_market_features
```

```python
df_market_features.to_csv('market_features.csv', index=False)
```
